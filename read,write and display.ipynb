```python
#load the binary image

import cv2

img=cv2.imread("im6.png",0)
ret,bw_img=cv2.threshold(img,127,300,cv2.THRESH_BINARY)
bw=cv2.threshold(img,300,127,cv2.THRESH_BINARY)
cv2.imshow("Binary",bw_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
